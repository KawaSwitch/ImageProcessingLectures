cv2.CannyによってCanny法によるエッジ検出を実装した

```
def showCannyEdges(img_path, minVal=100, maxVal=200, aperture_size=3, L2gradient=False, withOriginal=True):
  img = cv2.imread(img_path)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  img_gry = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
  edges = cv2.Canny(img_gry, minVal, maxVal)

  plt.figure(figsize=(12, 6))
  if withOriginal:
    plt.subplot(121),plt.imshow(img,cmap = 'gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])
  plt.subplot(122),plt.imshow(edges,cmap = 'gray')
  plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
```
